```python
# PN Table の読み込み
import pandas as pd

pn_table = pd.read_csv('./pn_ja.dic.txt', 
                    sep=':',
                    encoding='shift-jis',
                    names=('Word','Reading','POS', 'PN')
                      )


# PN Tableをデータフレームからdict型に変換しておく
word_list = list(pn_table['Word'])
pn_list = list(pn_table['PN'])  # 中身の型はnumpy.float64
pn_table_dic = dict(zip(word_list, pn_list))

#print(pn_table_dic)
```

```python
# 日本語評価極性辞書(JPSD) の読み込み
import pandas as pd

jpsd = pd.read_csv('./pn.csv.m3.120408.txt',
                    sep='\t',
                    encoding='utf-8',
                    names=('Word','descretePN','RULE')
                   )
jpsd = jpsd.iloc[:, 0:2]

# descretePN列のeを0,pを0.3,nを-0.3に置換しておく
jpsd["descretePN"] = jpsd["descretePN"].replace("e",0).replace("p",0.3).replace("n",-0.3)


# 日本語評価極性辞書をデータフレームからdict型に変換しておく
word_list = list(jpsd['Word'])
pn_list = list(jpsd['descretePN'])# 中身の型はnumpy.float64
jp_dict = dict(zip(word_list, pn_list))
# あとで使えるように元のdescretePNをリスト化しておく
descrete_pn = list(jp_dict.values())

for n,i in enumerate(descrete_pn):
    if i == 0.3:
        descrete_pn[n] = 1
    elif i == -0.3:
         descrete_pn[n] = -1
        
    
        
# pn_tableにもある名詞であり、かつPN値の符号が揃っていて、日本語評価極性辞書のPN値が0でない場合のみ置き換える
for jp_w in jp_dict.keys():
    if pn_table_dic.get(jp_w, None) != None and jp_dict[jp_w] * pn_table_dic[jp_w] > 0: # .get(x,y)関数は辞書のキーにxがなかった場合、yを出力する
        jp_dict[jp_w] = pn_table_dic[jp_w]
        
    elif pn_table_dic.get(jp_w, None) != None and jp_dict[jp_w] * pn_table_dic[jp_w] < 0:
        jp_dict[jp_w] = -1 * pn_table_dic[jp_w]
                
print(jp_dict)
```

```python
# jp_dictを2つのリスト型に変換
word = jp_dict.keys() 
pn = jp_dict.values()

# データフレームを作成
new_ipsd = pd.DataFrame()
# 列の追加
# new_ipsd["descretePN"] = descrete_pn
new_ipsd["Word"] = word 
new_ipsd["PN"] = pn

print(new_ipsd)
#new_ipsd.to_csv("noun_pn.csv")
```

```python
import numpy as np

verb = pd.read_csv('./verb.csv',
                    sep='\t',
                    encoding='utf-8',
                 )

verb = verb.rename(columns={"verb":"Word"}) # カラム名を入れ替え　

verb.index = np.arange(13307, 18587)
#print(verb)

verb = verb.ix[:,['Word','PN','descretePN']]  # 列の並び替え
verb = verb.iloc[:, 0:2] # 列の参照
#print(verb)

df_word = pd.concat([new_ipsd, verb], axis=0)
print(df_word)

df_word.to_csv("pn_dic.csv")
```
