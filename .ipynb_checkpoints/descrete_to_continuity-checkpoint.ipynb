```python
# モジュールのインポート
import re
import csv
import time
import pandas as pd
import matplotlib.pyplot as plt
import MeCab
import random


# 乾岡崎研究所の離散評価辞書（用言:verb）の読み込み
dic_df = pd.read_csv('wago.121808.pn',sep='\t',names=('descretePN','verb') ,encoding='utf-8')

# PN Tableを読み込み
pn_df = pd.read_csv('pn_ja.dic.txt',
                    sep=':',
                    encoding='utf-8',
                    names=('Word','Reading','POS', 'PN')
                   )
#MeCab辞書は標準に設定
m = MeCab.Tagger('')
```

```python
# PN Tableをデータフレームからdict型に変換しておく
word_list = list(pn_df['Word'])
pn_list = list(pn_df['PN'])  # 中身の型はnumpy.float64
pn_dict = dict(zip(word_list, pn_list))
```

```python
# -----テキストを形態素解析して辞書のリストを返す関数----- #
def get_diclist(verb):
    parsed = m.parse(verb)      # 形態素解析結果（改行を含む文字列として得られる）
    lines = parsed.split('\n')  # 解析結果を1行（1語）ごとに分けてリストにする
    lines = lines[0:-2]         # 後ろ2行は不要なので削除
    diclist = []
    for word in lines:
        l = re.split('\t|,',word)  # 各行はタブとカンマで区切られてるので
        d = {'Surface':l[0], 'POS1':l[1], 'POS2':l[2], 'BaseForm':l[7]}
        diclist.append(d)
    return(diclist)

#example
print(get_diclist("堪忍袋 の 緒 が 切れる"))
```

```python
# 形態素解析結果の単語ごとdictデータにPN値を追加する関数(入力は各単語の上記辞書型のリスト)
def add_pnvalue(diclist_old):
    diclist_new = []
    for word in diclist_old:
        base = word['BaseForm']        # 個々の辞書から基本形を取得
        if base in pn_dict:
            pn = float(pn_dict[base])  # 中身の型があれなので
        else:
            pn = 'notfound'            # その語がPN Tableになかった場合
        word['PN'] = pn
        diclist_new.append(word)
    return(diclist_new)
#example
vdic = [{'Surface': '堪忍袋','POS1': '名詞', 'POS2': '一般','BaseForm': '堪忍袋'},
        {'Surface': 'の', 'POS1': '助詞', 'POS2': '連体化', 'BaseForm': 'の'},
        {'Surface': '緒', 'POS1': '名詞', 'POS2': '一般', 'BaseForm': '緒'},
        {'Surface': 'が', 'POS1': '助詞', 'POS2': '格助詞', 'BaseForm': 'が'},
        {'Surface': '切れる', 'POS1': '動詞', 'POS2': '自立', 'BaseForm': '切れる'}]
print(add_pnvalue(vdic))
```

```python
#ごく普通にリストの値の平均値を返す関数
def mean(li):
    return sum(li)/len(li)

# 各用言のPN平均値をとる関数（乾岡崎研究所の辞書の用言が、複合体の形を取ることが多いため
def get_pnmean(diclist):
    pn_list = []
    for word in diclist:
        pn = word['PN']
        if pn != 'notfound':
            pn_list.append(pn)  # notfoundだった場合は追加もしない            
    if len(pn_list) > 0:        # 「全部notfound」じゃなければ
        pnmean = mean(pn_list)
    else:
        pnmean = 0              # 全部notfoundならゼロにする
    return(pnmean)
```

```python
pnmeans_list = []
for word in dic_df['verb']:
    diclist = get_diclist(str(word))
    diclist_new = add_pnvalue(diclist)
    pnmean = get_pnmean(diclist_new)
    pnmeans_list.append(pnmean)
```

```python
new_df = pd.DataFrame({'descretePN':dic_df['descretePN'],
                       'verb':dic_df['verb'],
                        'PN':pnmeans_list,
                       },
                       columns=['descretePN', 'verb', 'PN']
                      )
# PN値の降順でソート
new_df = new_df.sort_values(by='PN', ascending=False)
print(new_df)
```

```python
new_df.to_csv('new.csv',
              sep=",",
                index=None,
                encoding='utf-8',
                quoting=csv.QUOTE_NONNUMERIC
               )
```
